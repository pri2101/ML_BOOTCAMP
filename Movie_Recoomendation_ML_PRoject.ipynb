```python
import numpy as np
import pandas as pd
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
columns_name=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep="\t",names=columns_name)
df.head()
```

```python
df['item_id'].nunique()
```

```python
movies_title=pd.read_csv('u.item',sep="\|",header=None,encoding="ISO-8859-1")
```

```python
movies_title.shape
```

```python
movies_titles=movies_title[[0,1]]
movies_titles.columns=["item_id","title"]
movies_titles.head()
```

```python
df=pd.merge(df,movies_titles,on="item_id")
```

```python
df
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
moviemat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
moviemat.tail()
```

```python
youn_user_ratings=moviemat['Young Guns (1988)']
```

```python
youn_user_ratings.head(10)
```

```python
similar_to_youn=moviemat.corrwith(youn_user_ratings)
```

```python
similar_to_youn
```

```python
corr_youn=pd.DataFrame(similar_to_youn,columns=['correlation'])
```

```python
corr_youn.dropna(inplace=True)
```

```python
corr_youn
```

```python
corr_youn.sort_values('correlation',ascending=False).head(25)
```

```python
corr_youn=corr_youn.join(ratings['num of ratings'])
```

```python
corr_youn
```

```python
corr_youn[corr_youn['num of ratings']>100].sort_values('correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_movies=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_movies,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False)
    
    return predictions
```

```python
predict_my_movies=predict_movies('Secrets & Lies (1996)')
```
